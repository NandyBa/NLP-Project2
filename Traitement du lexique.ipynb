```python
import pandas as pd

from nltk.corpus import stopwords

from sklearn.feature_extraction.text import TfidfVectorizer

import spacy
# Load Our Model & NLP(TALN) Object
nlp = spacy.load('fr_core_news_sm')
```

```python
df = pd.read_csv('avisassurance_train.csv', sep=";")
```

```python
def enlever_les_verbes(nlp, avis):
    avis_sans_verbes = ''
    
    doc = nlp(avis)
    
    for token in doc:
        if token.pos_ != 'VERB':
            avis_sans_verbes += token       
    return avis_sans_verbes

def groupes_de_mots_les_plus_employés_par_note(df, note):
    stop_words_list = stopwords.words("french")
    model = TfidfVectorizer(ngram_range=(1,2), stop_words=stop_words_list)
    model = model.fit([ str(avis) for avis in df[df['note']==note]['avis'].to_list()])
    return model.vocabulary_
```

```python
groupes_de_mots_les_plus_employés_par_note(df, 1)
```
